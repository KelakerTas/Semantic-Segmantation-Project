```python
import os

data_fucus = os.listdir("Dataset/Fucus")
data_furcellaria_lumbricalis = os.listdir("Dataset/Furcellaria lumbricalis")
data_zostera_marina = os.listdir("Dataset/Zostera Marina")
```

```python
i = 0
for data in data_zostera_marina:
    i=i+1
    os.rename("Dataset/Zostera Marina/"+data,"Dataset/Zostera Marina/" +"Zostera Marina_" + str(i) + ".jpg")
```

```python
i = 0
for data in data_furcellaria_lumbricalis:
    i=i+1
    os.rename("Dataset/Furcellaria lumbricalis/"+data,"Dataset/Furcellaria lumbricalis/" +"Furcellaria lumbricalis_" + str(i) + ".jpg")
```

```python
import cv2

for species in os.listdir("Dataset"):
    
    path = "Dataset/" + species
    
    for spe in os.listdir(path):
        
        d = path + "/" + spe
        print(d)
        
        img = cv2.imread(d)
        
        dim = (1024,1024)
    
        resized_img = cv2.resize(img,dim, interpolation = cv2.INTER_AREA)
        
        cv2.imwrite("all_images/" + spe, resized_img)
        
```
